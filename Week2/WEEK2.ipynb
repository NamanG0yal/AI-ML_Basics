```python
from statsmodels.tools.sm_exceptions import  ValueWarning
import itertools
import datetime
!pip install prophet
from prophet import Prophet
import plotly.offline as py
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder,Normalizer
data=pd.read_csv('/content/train.csv')
data=data.iloc[:,1:]
le= LabelEncoder()
n=Normalizer()
A=data.iloc[:,:-1]
B=pd.DataFrame(map(float,data.iloc[:,-1]))
#Encoding of categorial data
for i in range(len(A.columns)):
  k=A.iloc[:,i]
  k=le.fit_transform(k)
  A.iloc[:,i]=k
#Normalisation of data
for column in A.columns:
  if A[column].abs().max()!=0:
    A[column] = A[column]/A[column].abs().max()
#More data processing
p = pd.concat([data.iloc[:,1], data.iloc[:,16]], axis=1, ignore_index=False)
p['Order Date']=pd.DatetimeIndex(p['Order Date'])
p.index.asfreq='M'
agg_functions = {'Order Date': 'first', 'Sales': 'sum'}
p=p.groupby(p['Order Date']).aggregate(agg_functions)
p = p.rename(columns={'Order Date': 'ds','Sales': 'y'})
#SARIMA use
result = adfuller(p['y'])
print('p-value: %f' % result[1])
#p-value < 0.05, hence stationary data
pa = d = q = range(0, 2)
pdq = list(itertools.product(pa, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(pa, d, q))]
k=p.set_index(['ds'])
for param in pdq:
  try:
    for param_seasonal in seasonal_pdq:
      mod = sm.tsa.statespace.SARIMAX(k,order=param,seasonal_order=param_seasonal,enforce_stationarity=False,enforce_invertibility=False)
      results = mod.fit()
      print('ARIMA{}x{}12 - AIC:{}'.format(param,param_seasonal,results.aic))
  except:
    continue
#0,1,1 ---- 0,1,1,12 has the lowest AIC---> used in the model
mod = sm.tsa.statespace.SARIMAX(k,order=(0, 1, 1),seasonal_order=(0, 1, 1, 12),enforce_stationarity=False,enforce_invertibility=False)
results = mod.fit()
print(results.summary().tables[1])
results.plot_diagnostics()
plt.show()
pred_uc = results.get_forecast(steps=20)
pred_ci = pred_uc.conf_int()
ax = k.plot(label='observed')
pred_uc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index,pred_ci.iloc[:, 0],pred_ci.iloc[:, 1])
ax.set_xlabel('Date')
ax.set_ylabel('Sales')
plt.legend()
plt.show()
#Prophet use
p_model=Prophet()
p_model.add_country_holidays('US')
p_model.fit(p)
future_dates = p_model.make_future_dataframe( periods=8, freq='M',include_history=True)
forecast = p_model.predict(future_dates)
print(forecast[['ds', 'yhat']])
p.plot()
plt.show()
p_model.plot(forecast)
```
